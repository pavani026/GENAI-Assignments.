```python
# Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))
            if amount < 50:
                print("Recharge amount must be at least ₹50. Try again.")
                continue
            if amount not in valid_plans:
                print(f"Invalid plan. Valid plans are: {valid_plans}. Try again.")
                continue
            print(f"Recharge of ₹{amount} successful!")
            break
        except ValueError:
            print("Please enter a valid number.")

# Run function
validate_recharge()
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

stock = {"Apple": 20, "Banana": 10, "Orange": 5, "Grapes": 30}
check_inventory(stock)
```

```python
#Problem 3: Student Result Processing System
def student_result(marks_list):
    total = 0
    for mark in marks_list:
        total += mark
    average = total / len(marks_list)
    print(f"Average Marks: {average}")
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

result = student_result([60, 45, 70, 80])
print(result)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare():
    base_fare = 50
    per_km = 12

    while True:
        try:
            distance = float(input("Enter distance in km: "))
            peak = input("Is it peak hour? (yes/no): ").lower()
            fare = base_fare + (per_km * distance)
            if peak == "yes":
                fare *= 1.25
            print(f"Total Fare: ₹{fare:.2f}")
            break
        except ValueError:
            print("Invalid input, please enter numbers.")
calculate_fare()
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def attendance_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = attendance_list.count("P")
    percentage = (present_days / total_days) * 100
    print(f"Attendance Percentage: {percentage:.2f}%")
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

result = attendance_eligibility(["P","A","P","P","A","P","P","P"])
print(result)
```

```python
#Problem 6: Password Strength Checker
def password_strength(password):
    special_chars = "@#$"
    if len(password) < 8:
        return "Weak: Less than 8 characters"
    
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in special_chars for char in password)

    if has_digit and has_special:
        return "Strong Password"
    elif not has_digit:
        return "Weak: Must contain at least one digit"
    else:
        return "Weak: Must contain at least one special character (@#$)"

print(password_strength("Pass123"))
print(password_strength("Pass@123"))
```
