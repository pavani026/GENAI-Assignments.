## Day 5

```python
def classify_heart_rate(heart_rate, activity_status):
    # check if the patient is at rest
    if activity_status == 'rest':
        if heart_rate < 60:
            status = "low"
            risk = "Possible bradycardia risk"
        elif heart_rate <= 100:
            status = "normal"
            risk = "No immediate risk"
        else:
            status = "high"
            risk = "Possible tachycardia risk"
    
    # check if patient is active
    elif activity_status == "active":
        if heart_rate < 60:
            status = "low"
            risk = "Unusually low for activity"
        elif heart_rate <= 100:
            status = "normal"
            risk = "Healthy active range"
        else:
            status = "high"
            risk = "Elevated heart rate during activity"
    
    else:
        return "Invalid activity status"
    
    # return final result
    return status, risk


# user inputs
heart_rate = int(input("Enter patient's heart rate: "))
activity_status = input("Enter activity status (rest/active): ").lower()

# function call
result = classify_heart_rate(heart_rate, activity_status)
print("Status:", result[0])
print("Risk:", result[1])
```

## Crop yield Analysis

```python
def analyze_yield(yields):
    total = 0
    # calculate total yield
    for value in yields:
        total += value
    
    # calculate average
    average = total / len(yields)
    
    # decide yield status
    if average >= 50:
        status = "Good production"
    else:
        status = "Bad production"
    
    # return all results
    return total, average, status


# empty list
daily_yields = []

# ask days 
days = int(input("Enter the number of days: "))
for i in range(days):
    yield_value = int(input(f"Enter yield for day {i+1} (kg): "))
    daily_yields.append(yield_value)

# call function 
total, average, result = analyze_yield(daily_yields)

# output
print("Daily yields:", daily_yields)
print("Total:", total)
print("Average:", average)
print("Status:", result)

```
